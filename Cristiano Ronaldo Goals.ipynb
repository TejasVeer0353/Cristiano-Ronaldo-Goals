# EDA, Data Visualization¶

```python

plt.figure(figsize=(10, 3))
plt.xticks(fontsize=10,rotation='vertical')
sns.countplot(x=df['Competition'], order=df.Competition.value_counts().sort_values
              (ascending=False).index)
plt.title('Competition played')
```

The above data shows that Ronaldo has his highest appearance in La-Liga matches than any other league or cup in his career, this makes sense due to the time he spent at Real madrid club






```python

plt.figure(figsize=(10, 3))
plt.xticks(fontsize=10, rotation='vertical')
sns.countplot(x=df['Club'], order= df.Club.value_counts().sort_values
              (ascending=False).index)
plt.title('Matches appearances per club')
```

Since ronaldo has his most appearances in Madrid, it makes sense that most of his mathes are in the club

```python

plt.figure(figsize = (10, 3))
plt.xticks(fontsize = 10, rotation = 'vertical')
sns.countplot(x=df.Playing_Position,order=df.Playing_Position.value_counts().sort_values(ascending=False).index)
plt.title('Ronaldo playing positions')
```

Ronaldo plays more on the left winger position, logically, this will means that he likely scores with his right foot, let's find that out below.



```python

plt.figure(figsize= (10, 3))
plt.xticks(fontsize = 10, rotation = 'vertical')
sns.countplot(x=df.Type, order = df.Type.value_counts().sort_values
              (ascending = False).index)
plt.title('Ronaldo scoring types')
```

Since Ronaldo plays the left wing position, scoring most of his goals with right foot makes more sense. But interesingly, this guy has scored over 100 goals with his left foot and head, including penalty!



```python

plt.figure(figsize=(20,5))
plt.xticks(fontsize=8, rotation = 'vertical')
sns.countplot(x=df.Opponent, order = df.Opponent.value_counts().sort_values
             (ascending= False).index )
plt.title('Ronaldo highest opponent')
```

Ronaldo has played most of his matches against spanis clubs, the highest being sevila, Altletico madird. Yea, I know most people think it's Barcelona, well, it's not!

```python

plt.figure(figsize=(15, 5))
plt.xticks(fontsize=10, rotation = 'vertical')
sns.countplot(x=df.Goal_assist, order=df.Goal_assist.value_counts().sort_values(ascending = False).index)
plt.title('Ronaldo highest goal assistant!')
```

Having scored most of his goal in Madrid, Ronaldo's Biggest assistant is Karim Benzema

```python

pk =df.loc[df.Type == 'Penalty']
plt.figure(figsize=(15, 5))
plt.xticks(fontsize=10, rotation = 'vertical')
sns.countplot(x=pk.Competition, order = pk.Competition.value_counts().
             sort_values(ascending = False).index)
plt.title('Penalty per competition')
```

Ronaldo played his highest penalty in LaLiga. followed by Seria A and Champions league.

```python

plt.figure(figsize=(18, 5))
plt.xticks(fontsize=10, rotation ='vertical')
sns.countplot(x=df.Minute, order = df.Minute.value_counts().
             sort_values(ascending = False).index)
plt.title("Number of minutes played in Ronaldo's career")
```

Ronaldo has played more full-time matches



```python

plt.figure(figsize=(18, 5))
plt.xticks(fontsize=10, rotation ='vertical')
sns.countplot(x=df.Matchday,  order = df.Matchday.value_counts().
             sort_values(ascending = False).index)
plt.title('Number of match types')
```
